The primary purpose of recommendation engines is to provide personalized suggestions or recommendations to users, helping them discover new items or content that align with their tastes.

building model

```python
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

# Find the best value of k using cross-validation
best_k = None
best_score = float('-inf')

for k in range(1, 7):  # Try different values of k
    knn = KNeighborsClassifier(n_neighbors=k)
    k_folds=KFold(n_splits=10)
    scores = cross_val_score(knn, X_train, y_train, cv=k_folds)  # Perform cross-validation
    avg_score = np.mean(scores)

    if avg_score > best_score:
        best_score = avg_score
        best_k = k

# Train the model using the best k on the entire training set
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

#Training accuracy
train_accuracy = best_knn.score(X_train, y_train)
print("\n Training Accuracy:", train_accuracy)

#Testing accuracy
test_accuracy = best_knn.score(X_test, y_test)
print("\n Testing Accuracy:", test_accuracy)

```
